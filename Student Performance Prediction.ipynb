# Student Performance Prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

The student performance detection project aims to predict students' academic performance based on various input features, such as the number of hours studied, previous scores, extracurricular activities, sleep hours, and sample question papers practiced. Using linear regression, the model is trained to learn the relationships between these factors and the performance index, allowing it to make predictions on new students' performance given their input data.

```python
data=pd.read_csv("Student_Performance.csv")
X = data[['Hours Studied', 'Previous Scores', 'Sleep Hours','Sample Question Papers Practiced']]
y = data['Performance Index']
X = X.values
y = y.values
print("Input features: ", X)
print("Target Features: ",y)
```

```python
print ('The shape of input is:', X.shape)
print ('The shape of target variable is: ', y.shape)
print ('Number of training examples (m):', len(X))
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 13))
fig.suptitle('Student Performance vs. Input Features')

# Plot each input feature against the target variable (Performance Index)
axs[0][0].scatter(X[:, 0], y)  # Accessing the first column of X using index 0
axs[0][0].set_xlabel('Hours Studied')
axs[0][0].set_ylabel('Performance Index')

axs[0][1].scatter(X[:, 1], y)  # Accessing the second column of X using index 1
axs[0][1].set_xlabel('Previous Scores')
axs[0][1].set_ylabel('Performance Index')

axs[1][0].scatter(X[:, 2], y)  # Accessing the fourth column of X using index 3
axs[1][0].set_xlabel('Sleep Hours')
axs[1][0].set_ylabel('Performance Index')

axs[1][1].scatter(X[:, 3], y)  # Accessing the fifth column of X using index 4
axs[1][1].set_xlabel('Sample Question Papers Practiced')
axs[1][1].set_ylabel('Performance Index')

# Adjust the layout to prevent overlapping of subplots
plt.tight_layout()

# Show the plot
plt.show()
```

Cost Function

```python
def compute_cost(X, y, w, b): 
    
    m = X.shape[0]
    cost = 0.0
    for i in range(m):                                
        f_wb_i = np.dot(X[i], w) + b       
        cost = cost + (f_wb_i - y[i])**2       
    cost = cost / (2 * m)                          
    return cost
```

Gradient Descent

```python
def compute_gradient(X, y, w, b): 

    m,n = X.shape          
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):                             
        err = (np.dot(X[i], w) + b) - y[i]   
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * X[i, j]    
        dj_db = dj_db + err
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m                                   
    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    
    w = w_in 
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_db,dj_dw = gradient_function(X, y, w, b)  
        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               
        b = b - alpha * dj_db              
        print(f"After {i} iteration cost function = {cost_function(X,y,w,b)}")
    return w, b
```

```python
# initialize parameters
initial_w =np.random.randn(4)
initial_b = 0
iterations = 2000
alpha = 0.0001
# run gradient descent 
w_final, b_final = gradient_descent(X,y, initial_w, initial_b,
                                                    compute_cost, compute_gradient, 
                                                    alpha, iterations)
```

```python
print("w= ",w_final)
print("b= ",b_final)
```

```python
# Calculate predicted performance index based on the model's parameters
predicted_y = np.dot(X, w_final) + b_final

fig, axs = plt.subplots(2, 2, figsize=(12, 13))
fig.suptitle('Student Performance vs. Input Features')

# Plot each input feature against the target variable (Performance Index)
axs[0][0].scatter(X[:, 0], y, label='Actual')
axs[0][0].scatter(X[:, 0], predicted_y, color='red', marker='x', label='Predicted')
axs[0][0].set_xlabel('Hours Studied')
axs[0][0].set_ylabel('Performance Index')
axs[0][0].legend()

axs[0][1].scatter(X[:, 1], y, label='Actual')
axs[0][1].scatter(X[:, 1], predicted_y, color='red', marker='x', label='Predicted')
axs[0][1].set_xlabel('Previous Scores')
axs[0][1].set_ylabel('Performance Index')
axs[0][1].legend()

axs[1][0].scatter(X[:, 2], y, label='Actual')
axs[1][0].scatter(X[:, 2], predicted_y, color='red', marker='x', label='Predicted')
axs[1][0].set_xlabel('Sleep Hours')
axs[1][0].set_ylabel('Performance Index')
axs[1][0].legend()

axs[1][1].scatter(X[:, 3], y, label='Actual')
axs[1][1].scatter(X[:, 3], predicted_y, color='red', marker='x', label='Predicted')
axs[1][1].set_xlabel('Sample Question Papers Practiced')
axs[1][1].set_ylabel('Performance Index')
axs[1][1].legend()

# Adjust the layout to prevent overlapping of subplots
plt.tight_layout()

# Show the plot
plt.show()
```
